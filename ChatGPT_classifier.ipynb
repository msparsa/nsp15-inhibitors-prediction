```python
from typing import Callable
    
def parse_example(model: str, utterance: str, **kwargs: dict) -> str:
    """Return the semantic parse of the utterance"""
    response =  openai.ChatCompletion.create(
        model=model,
        messages=[
            {"role": "user", "content": utterance},
        ],
        temperature=0
)
    return response.choices[0].message.content
```

```python
import multiprocessing
from tqdm.notebook import tqdm
from multiprocessing import Pool

# Define the function to run in each process
def process_model(gpt_model):
    row = {}
    row['model'] = gpt_model
    try:
        parse = parse_example(model=gpt_model, utterance=query)
        row['response'] = parse
    except Exception as e:
        print(gpt_model, e)
        row['response'] = None
    return row

# List of models to process
models = ['gpt-4-turbo', 'gpt-4-turbo-2024-04-09', 'gpt-4-turbo-preview', 'gpt-4-0125-preview', 'gpt-4-1106-preview',
          'gpt-4-vision-preview', 'gpt-4-1106-vision-preview', 'gpt-3.5-turbo-0125', 'gpt-3.5-turbo', 'gpt-3.5-turbo-1106']

# Create a pool of workers equal to the number of available CPU cores
pool = Pool(10)

# Map the function over the models using the pool
results = list(tqdm(pool.imap(process_model, models), total=len(models)))

# Close the pool and wait for the work to finish
pool.close()
pool.join()

# Display the results
from IPython.display import display
display(results)
```
